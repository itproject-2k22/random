```python
import random
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def generateDataRandom():
    data = []
    for i in range(100):
        data.append(random.random())
    return data

def getMean(data):
    sum = 0
    for i in data:
        sum = sum + i
    mean = sum/len(data)
    return mean

def getVariance(data):
    sum = 0
    mean = getMean(data)
    for i in data:
        sum = sum + (i-mean)**2
    variance = sum/len(data)
    return variance

def getStd(data):
    std = getVariance(data)**0.5
    return std 

def getMin(data):
    min = data[0]
    for i in data:
        if min > i:
            min = i
    return min

def getMax(data):
    max = data[0]
    for i in data:
        if max < i:
            max = i
    return max

def getRange(data):
    max = getMax(data)
    min = getMin(data)
    range = max-min
    return range

def getHistogram(data, title='', xlabel='', ylabel=''):
    plt.hist(data, edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
def getPlot(data, title='', xlabel='', ylabel=''):
    plt.plot(data)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def getBoxplot(data, title='', xlabel='', ylabel=''):
    plt.boxplot(data)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
def normalTestPlot(data, title=''):
    stats.probplot(data, dist='norm', plot=plt)
    plt.title(title)
    plt.show
    
def getScatterPlot(x, y, title='', xlabel='', ylabel=''):
    plt.scatter(x,y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
data = generateDataRandom()
data2 = generateDataRandom()
print(data, end='\n')
print("mean\t:", getMean(data))
print("var\t:", getVariance(data))
print("std\t:", getStd(data))
print("min\t:", getMin(data))
print("max\t:", getMax(data))
print("range\t:", getMax(data))
getHistogram(data)
getPlot(data)
getBoxplot(data)
normalTestPlot(data)
getScatterPlot(data, data2)

```
