```python
import random
```

```python
def generateDataRandom():
    data = []
    for i in range(100):
        data.append(random.random())
    return data

def getMean(data):
    sum = 0
    for i in data:
        sum = sum + i
    mean = sum/len(data)
    return mean

def getVariance(data):
    sum = 0
    mean = getMean(data)
    for i in data:
        sum = sum + (i-mean)**2
    variance = sum/len(data)
    return variance
```

```python
data = generateDataRandom()
print(data)
```
