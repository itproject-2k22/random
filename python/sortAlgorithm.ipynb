```python
import random
```

```python
def generateDataRandom():
    data = []
    for i in range(100):
        data.append(random.random())
    return data
```

```python
def selectionSort(data, type='asc'):
    for i in range(len(data)):
        idx=i
        for j in range(i+1,len(data)):
            if type=='asc':
                if data[j] < data[idx]:
                    idx = j
            if type=='desc':
                if data[j] > data[idx]:
                    idx = j
        data[i], data[idx] = data[idx], data[i]
    return data
```

```python
data = generateDataRandom()
print('raw data\t:', data)
print('data asc\t:', selectionSort(data))
print('data des\t:', selectionSort(data,type='desc'))
```
